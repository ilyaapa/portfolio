# Анализ оттока клиентов банка

## Описание проекта

В банке «Метанпромбанк» последнее время значительно повысился отток клиентов, это начинает становиться проблемой, которая угрожает существованию банка. Привлекать новых клиентов дорого, дешевле удержать тех, которых банк уже знает и с которыми работает.

Отделу маркетинга банка необходимы результаты анализа оттока клиентов, которые покажут, какие клиенты уходят из банка, а так же позволит составить сегменты клиентов, которые склонны уходить из банка.

Информация нужна менеджеру, который занимается маркетиновыми кампаниями и работает в тесной связке с менеджерами продуктов.
Поскольку в отделе маркетинга нет автоматизированных систем рассылок и письма каждому клиенту пишутся вручную, необходимо максимально эффективно выделить сегменты отточных клиентов, чтобы они были однородными, высокоотточными и покрывали максимум проблемных зон банка. Таким образом, маркетинговое воздействие на эти сегменты будет максимально эффективным и должно будет привести к уменьшению оттока.

Также необходимо дать рекомендации по мероприятиям, которые можно провести, чтобы вернуть клиентов в банк или удержать сомневающихся от оттока.

Ранее уже проводилось подобное исследование, но его результаты устарели и по ним не возможно выяснить, в чём сейчас заключается проблема и как её можно решить. Кроме этого, состав клиентов банка изменился и старые сегменты больше не подходят.

## Описание данных

Датасет `bank_scrooge.csv` содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах Ростов Великий и Рыбинск.

- USERID — идентификатор пользователя,
- score — баллы кредитного скоринга,
- city — город,
- gender — пол,
- age — возраст,
- equity — приблизительная оценка собственности клиента,
- balance — баланс на счёте,
- products — количество продуктов, которыми пользуется клиент,
- credit_card — есть ли кредитная карта,
- last_activity — был ли клиент активен последнее время,
- EST_SALARY — оценочная заработная плата клиента,
- сhurn — признак оттока.

Не все названия столбцов в `snake_case`. Исправим это.

```python
bank_balance_isna = bank[bank['balance'].isna()]
bank_balance_isna.head(20)
```

```python
bank_balance_isna.drop('user_id', axis=1).describe()
```

81.6% клиентов с пропуском в столбце `balance` являются обладателями кредитных карт, а также почти у всех таких клиентов всего 1 продукт банка, соответственно это продукт - кредитная карта. Сравним клиентов с пропуском в столбце `balance` и без по среднему значению признаков `credit_card` и `products`. Для этого создадим новый признак, принимающий 1, если в наблюдении пропущено значение баланса, и 0, если в наблюдении значение баланса не пропущено.

```python
bank_balance_na = bank[['balance', 'score', 'age', 'equity', 'credit_card', 'products', 'last_activity', 'est_salary']]
bank_balance_na['balance_na'] = np.where(bank_balance_na['balance'].isnull(), 0,1)
bank_balance_na = bank_balance_na.drop('balance', axis=1)
bank_balance_na.groupby('balance_na').agg({'score':'mean',
                                           'age':'mean', 
                                           'equity':'mean',
                                           'credit_card':'mean',
                                           'products':'mean',
                                           'last_activity':'mean',
                                           'est_salary':'mean'})
```

Видим, что клиенты с пропуском значения баланса являются владельцами единственного продукта - кредитной карты, в то аремя, как клиенты без пропуска в среднем имеют по 2 продукта банка, что не исключает, что они также владеют кредитной картой (поскольку 64.0% пользователей без пропуска в значении баланса владеют кредитной картой), но кроме этого, у них есть второй продукт банка, связанный с наличием счета (вклад или дебетовая карта). Кроме этого, есть явное различие по оценке собственности клиента - у клиентов с пропуском значения баланса среднее значение этого признака 0.05, в то время, как у клиентов без пропуска это значение равно 3.4. Это может быть связано как с отсутствием или непредоставлением информации клиентами, так и с объективной оценкой собственности клиента или высокой закредитованностью, т.к. у клиента может быть кредит или кредитная карат не только в банке «Метанпромбанк». При этом средняя зарплата клиентов с пропуском значения баланса почти в 2 раза выше, чем у клиентов без пропуска - 226310.21 р. и 124501.87 р. соответственно. По остальным признакам - кредитному скорингу, возрату, активности в последнее время - между клиентами значимых отличий нет.

Пропуски в столбце `balance` появились не случайно, они зависят от других данных. Эти выбрасы также не являются ошибками, но их мало и они не оказывают серьезного влияния на результаты анализа. Для этих параметров расчитаем только средние значения.

Дополнительно расчитаем абсолютную и относительную разницы соответствующих значений отточных и неотточных клиентов.

```python
clients_portraits = bank.groupby(['churn']).agg({'balance':['mean', 'median'],
                                                 'est_salary':['mean', 'median'],
                                                 'score':'mean',
                                                 'age':'mean',
                                                 'products':'mean'})
clients_portraits.columns = ('balance_mean',
                             'balance_median',
                             'salary_mean',
                             'salary_median',
                             'score_mean',
                             'age_mean',
                             'products_mean')
clients_portraits = clients_portraits.transpose().reset_index()
clients_portraits.columns = ('parameter', 'hold_client', 'churn_client')
clients_portraits['delta'] = clients_portraits['churn_client'] - clients_portraits['hold_client']
clients_portraits['delta, %'] = ((clients_portraits['churn_client'] - clients_portraits['hold_client'])/
                                   clients_portraits['hold_client']) * 100
clients_portraits = clients_portraits.round(2)
clients_portraits
```

Для дискретных значений выведем моду.

```python
clients_portraits_disc = bank.groupby(['churn']).agg({'credit_card':pd.Series.mode,
                                                      'gender':pd.Series.mode})
clients_portraits_disc.columns = ('credit_card_mode',
                             'gender_code_mode')
clients_portraits_disc = clients_portraits_disc.transpose().reset_index()
clients_portraits_disc.columns = ('parameter', 'hold_client', 'churn_client')
clients_portraits_disc
```

## Проверка статистических гипотез

**Проверим гипотезу различия дохода между отточными и неотточными клиентами.**

Для этого воспользуемся методами проерки гипотезы о равенстве средних двух генеральных совокупностей с применением t-критерия Стьюдента  u-критерия Манна-Уитни.

Т-критерий Стьюдента обычно применяется для проверки равенства средних значений в двух выборках. Для применения данного критерия необходимо, чтобы выборочные средние имели нормальное распределение и данных не было значительного числа выбросов. 

U-критерий Манна-Уитни также является статистическим критерием, используемым для оценки различий между двумя независимыми выборками. Применяется при несоблюдении требований к данным для t-критерия Манна-Уитни. Позволяет выявлять различия в значении параметра между малыми выборками. Для применения данного критерия требуется, чтобы в выборочных данных не бытло совпадающих значений или таких совпадений бытло очень мало (до 10).

Сформулируем нулевую и альтернативную гипотезы:
- H0 - уровень дохода отточных и неотточных клиентов одинаковый.
- Н1 - уровень дохода отточных и неотточных клиентов разный.

Выборка зарплат отточных клиентов

```python
churn_salary = bank_churn['est_salary']
```

Среднее значение зарплаты неотточных клиентов

```python
hold_salary = bank_hold['est_salary']
```

Примем уровень статистической значимости равным 5%.

```python
alpha = 0.05 
```

Проверяем гипотезу с применением t-критерия Стьюдента.

```python
results = st.ttest_ind(
    churn_salary, 
    hold_salary)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: уровень дохода отточных и неотточных клиентов разный.')
else:
    print('Не отвергаем нулевую гипотезу: уровень дохода отточных и неотточных клиентов одинаковый.')
```

Прверяем гипотезу с применением u-критерия Манна-Уитни.

Для этого воспользуемся методами проерки гипотезы о равенстве средних двух генеральных совокупностей.

Сформулируем нулевую и альтернативную гипотезы:
- H0 - баланс отточных и неотточных клиентов одинаковый.
- Н1 - баланс отточных клиентов больше баланса неотточных клиентов.

Выборка балансов отточных клиентов

**Проверим гипотезу о равенстве доли мужчин и женщин среди отточных и неотточных клиентов.**

Для проверки гипотез о бинарных признаках будем использовать z-критерий двух пропорций, который основан на вычислении отношения разницы между случайной величиной и математическим ожиданием к стандартной ошибке этой случайной величины, т.е. показывает, является ли разница между двумя выборками статистически значмой. Для применения данного критерия необходимо, чтобы исходные данные имели нормальное распределение и была известна дисперсия генеральной совокупности. 

Сформулируем нулевую и альтернативную гипотезы (достаточно проверить гипотезу о равенстве доли мужчин среди отточных и неотточных клиентов):
- H0 - доля мужчин среди отточных и неотточных клиентов одинаковая.
- Н1 - доля мужчин среди отточных и неотточных клиентов разная.

Объявим функцию для проведения теста.

```python
def ztest(g1, g2, n1, n2, alpha):
    p1 = n1/g1
    p2 = n2/g2
    p_combined = (n1 + n2) / (g1 + g2)
    difference = p1 - p2
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/g1 + 1/g2))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу:  между долями пользователей в группах есть статистически значимая разница')
    else:
        print('Не можем отвергнуть нулевую гипотезу: между долями пользователей в группах нет статистически значимой разницы')
    print('p-value =', p_value, 'при alpha =', alpha)
```

Подготовим данные для проведения теста:
- количество отточных клиентов;
- количество неотточных клиентов;
- количество мужчин среди отточных клиентов;
- количество мужчин среди неотточных клиентов.

```python
bank_churn_clients = bank_churn['user_id'].count()
bank_hold_clients = bank_hold['user_id'].count()
bank_churn_clients_male = bank_churn['gender_code'].sum()
bank_hold_clients_male = bank_hold['gender_code'].sum()
```

Проведем тест.

```python
ztest(bank_churn_clients, bank_hold_clients, bank_churn_clients_male, bank_hold_clients_male, alpha)
```

Для проверки выведем доли мужчин среди отточных и неотточных клиентов.

```python
print('Доля мужчин среди отточных клиентов равна', round(bank_churn_clients_male/bank_churn_clients, 2))
print('Доля мужчин среди неотточных клиентов равна', round(bank_hold_clients_male/bank_hold_clients, 2))
```
